```python
import random
```

```python
def train_markov_chain(text):
    words = text.lower().split()
    markov_chain = {}

    for i in range(len(words) - 1):
        current_word = words[i]
        next_word = words[i + 1]

        if current_word not in markov_chain:
            markov_chain[current_word] = []

        markov_chain[current_word].append(next_word)

    return markov_chain
```

```python
def generate_text(markov_chain, start_word, length=20):
    current_word = start_word.lower()
    result = [current_word]

    for _ in range(length - 1):
        if current_word not in markov_chain:
            break
        next_word = random.choice(markov_chain[current_word])
        result.append(next_word)
        current_word = next_word

    return " ".join(result)
```

```python
# Sample training text
text = """
artificial intelligence is the future
artificial intelligence is transforming the world
machine learning is a part of artificial intelligence
"""

# Train model
markov_model = train_markov_chain(text)

# Generate text
generated_text = generate_text(markov_model, start_word="artificial", length=15)

print("Generated Text:")
print(generated_text)
```
